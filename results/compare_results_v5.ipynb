### **Análise dos resultados do grid search**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
# Load the csv file
df = pd.read_csv('results/grid_search/grid_search_results_v5.csv')
df.head()
```

```python
df.info()
```

```python
# Drop columns that are not needed
columns_to_drop = ['index', 'train_loss', 'train_accuracy', 'train_precision_macro_avg',\
                    'train_recall_macro_avg', 'train_f1_score_macro_avg',\
                    'val_accuracy', 'test_loss','test_accuracy', 'test_precision_macro_avg',\
                    'test_recall_macro_avg', 'test_f1_score_macro_avg']

df.drop(columns=columns_to_drop, axis=1, inplace=True)
```

```python
# Rename the columns taking out the "val"
new_columns = { 'val_loss' : 'loss',\
                'val_precision_macro_avg': 'precision_macro_avg', \
                'val_recall_macro_avg': 'recall_macro_avg', \
                'val_f1_score_macro_avg': 'f1_score_macro_avg'}

df = df.rename(columns=new_columns)
```

```python
df.insert(6, 'time', '')
```

```python
# Uncomment to show the first 10 rows of the dataframes
df.info()
# df.head(10)
```

```python
# Iterate over the rows of the dataframe
for index, row in df.iterrows():
    # Get the start and end time in ISO format
    start_datetime = datetime.fromisoformat(df.at[index, 'time_start'])
    fim_datetime = datetime.fromisoformat(df.at[index, 'time_end'])

    # Calculate the time
    time = fim_datetime - start_datetime

    # Get the hours, minutes and seconds
    hours, remainder = divmod(time.seconds, 3600)
    minutes, seconds = divmod(remainder, 60)

    # Format the time as 'HH:MM:SS'
    formatted_time = f"{hours:02d}:{minutes:02d}:{seconds:02d}"

    # Add the formatted time to the dataframe
    df.at[index, 'time'] = formatted_time
```

```python
df.head()
```

```python
# Divide the df into 2 DFs based on the model name
df_rajpurkar = df.query('model_name == "rajpurkar"')
df_ribeiro = df.query('model_name == "ribeiro"')
```

```python
# Delete the columns model_name
df_rajpurkar.drop(columns='model_name', axis=1, inplace=True)
df_ribeiro.drop(columns='model_name', axis=1, inplace=True)
```

```python
# Reset the index
df_rajpurkar = df_rajpurkar.reset_index(drop=True)
df_ribeiro = df_ribeiro.reset_index(drop=True)
```

```python
# df_rajpurkar.info()
# df_ribeiro.info()

df_rajpurkar.head(5)
# df_ribeiro.head(5)
```

```python
# Get the pivot_table for f1 score of each model
f1_rajpurkar = pd.pivot_table(df_rajpurkar, values = 'f1_score_macro_avg',\
                                index = ['batch_size'], columns = ['optimizer', 'learning_rate'])

f1_ribeiro = pd.pivot_table(df_ribeiro, values = 'f1_score_macro_avg',\
                                index = ['batch_size'], columns = ['optimizer', 'learning_rate'])
```

```python
# Get the pivot_table for the val_loss of each model
loss_rajpurkar = pd.pivot_table(df_rajpurkar, values = 'loss',\
                                index = ['batch_size'], columns = ['optimizer', 'learning_rate'])

loss_ribeiro = pd.pivot_table(df_ribeiro, values = 'loss',\
                                index = ['batch_size'], columns = ['optimizer', 'learning_rate'])
```

Análise dos resultados de F1-score

```python
# Get the index of highest f1_score_macro_avg from the df
ind_max_raj = df_rajpurkar['f1_score_macro_avg'].idxmax()
ind_max_rib = df_ribeiro['f1_score_macro_avg'].idxmax()
```

```python
# Get the highest f1_score_macro_avg from the df
f1_max_raj = df_rajpurkar['f1_score_macro_avg'].max()
f1_max_rib = df_ribeiro['f1_score_macro_avg'].max()

print(f'Max F1-score from Rajpurkar = {f1_max_raj:.3f}')
print(f'Max F1-score from Ribeiro = {f1_max_rib:.3f}')
```

```python
# Get the version, batch_size, optimizer and learning_rate from the highest f1_score_macro_avg
batch_max_raj = df_rajpurkar.loc[ind_max_raj,'batch_size']
opt_max_raj = df_rajpurkar.loc[ind_max_raj,'optimizer']
lr_max_raj = df_rajpurkar.loc[ind_max_raj,'learning_rate']
loss_raj = df_rajpurkar.loc[ind_max_raj,'loss']

batch_max_rib = df_ribeiro.loc[ind_max_rib,'batch_size']
opt_max_rib = df_ribeiro.loc[ind_max_rib,'optimizer']
lr_max_rib = df_ribeiro.loc[ind_max_rib,'learning_rate']
loss_rib = df_ribeiro.loc[ind_max_rib,'loss']
```

```python
print(f"Rajpurkar's best model: \nF1-score = {f1_max_raj:.3f}, loss = {loss_raj:.3f},\
       \nbatch_size = {batch_max_raj}, optimizer = {opt_max_raj}, learning_rate = {lr_max_raj}\n")

print(f"Ribeiro's best model: \nF1-score = {f1_max_rib:.3f}, loss = {loss_rib:.3f},\
       \nbatch_size = {batch_max_rib}, optimizer = {opt_max_rib}, learning_rate = {lr_max_rib}")
```
